```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

print("Libraries imported successfully!")
```

## **Data Loading**

```python
try:
    df = pd.read_csv('UPI-Dataset.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: 'UPI-Dataset.csv' not found. Please make sure you have uploaded the file to your Colab session.")
print("\nOriginal column names:")
print(df.columns.tolist())

original_cols = df.columns.tolist()

import re
def clean_col_names(df):
    cols = df.columns
    new_cols = []
    for col in cols:
        new_col = col.lower()
        new_col = re.sub(r'[^a-zA-Z0-9_]+', '_', new_col)
        new_col = new_col.strip('_')
        new_cols.append(new_col)
    df.columns = new_cols
    return df

df = clean_col_names(df)

print("\nStandardized column names:")
print(df.columns.tolist())
print("\n--- Column Name Changes ---")
for orig, new in zip(original_cols, df.columns):
    if orig != new:
        print(f"'{orig}'  --->  '{new}'")
print("\nFirst 5 rows of the dataset:")
display(df.head())
print("\nDataset Info:")
df.info()
print("\nDescriptive Statistics:")
display(df.describe())
```

## Data Cleaning and Feature Engineering

```python
print("Missing values per column before cleaning:")
print(df.isnull().sum())
df.dropna(inplace=True)
print("\nMissing values after dropping NaN rows:", df.isnull().sum().sum())


# --- Feature Engineering ---
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['transaction_hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()
df['transaction_month'] = df['timestamp'].dt.month_name()

print("\nNew temporal features created successfully:")
display(df[['timestamp', 'transaction_hour', 'day_of_week', 'transaction_month']].head())
```

## **Univariate Analysis**

```python
plt.figure(figsize=(14, 6))
sns.histplot(df['amount_inr'], bins=100, kde=True)
plt.title('Distribution of Transaction Amount (INR)', fontsize=16)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xscale('log')
plt.show()

plt.figure(figsize=(8, 6))
ax = sns.countplot(x='fraud_flag', data=df, palette='pastel')
plt.title('Distribution of Fraudulent vs. Non-Fraudulent Transactions', fontsize=16)
plt.xlabel('Is Fraud? (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.show()

# Transactions by Hour of Day
plt.figure(figsize=(14, 7))
sns.countplot(x='transaction_hour', data=df, palette='viridis')
plt.title('Total Transactions by Hour of Day', fontsize=16)
plt.xlabel('Hour of Day (0-23)', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.show()
```

## **Bivariate Analysis**

```python
# 1 Transaction Amount vs Fraud Status
plt.figure(figsize=(10, 7))
sns.boxplot(x='fraud_flag', y='amount_inr', data=df, palette='coolwarm')
plt.title('Transaction Amount Distribution by Fraud Status', fontsize=16)
plt.xlabel('Is Fraud?', fontsize=12)
plt.ylabel('Transaction Amount (INR)', fontsize=12)
plt.yscale('log')
plt.show()


plt.figure(figsize=(14, 7))
sns.countplot(x='transaction_hour', hue='fraud_flag', data=df, palette='RdYlGn')
plt.title('Fraud vs. Non-Fraud Transactions by Hour of Day', fontsize=16)
plt.xlabel('Hour of Day (0-23)', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.legend(title='Is Fraud?', labels=['No', 'Yes'])
plt.show()

plt.figure(figsize=(12, 7))
sns.countplot(x='device_type', hue='fraud_flag', data=df, palette='magma')
plt.title('Fraud vs. Non-Fraud Transactions by Device Type', fontsize=16)
plt.xlabel('Device Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Is Fraud?', labels=['No', 'Yes'])
plt.show()
```

##  Network Analysis with NetworkX

```python
fraud_sender_banks = df[df['fraud_flag'] == 1]['sender_bank'].unique()
fraud_receiver_banks = df[df['fraud_flag'] == 1]['receiver_bank'].unique()

fraud_related_banks = set(fraud_sender_banks) | set(fraud_receiver_banks)
subgraph_df = df[(df['sender_bank'].isin(fraud_related_banks)) | (df['receiver_bank'].isin(fraud_related_banks))]

print(f"Original dataset shape: {df.shape}")
print(f"Subgraph dataset shape for visualization: {subgraph_df.shape}")
G = nx.from_pandas_edgelist(subgraph_df,
                           source='sender_bank',
                           target='receiver_bank',
                           edge_attr='fraud_flag',
                           create_using=nx.DiGraph())
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(G, k=0.8, iterations=50)
node_colors = ['red' if node in fraud_related_banks else 'skyblue' for node in G.nodes()]
edge_colors = [ 'red' if G.edges[edge]['fraud_flag'] == 1 else 'lightgray' for edge in G.edges()]
edge_widths = [ 2.5 if G.edges[edge]['fraud_flag'] == 1 else 0.5 for edge in G.edges()]
edge_alphas = [ 1.0 if G.edges[edge]['fraud_flag'] == 1 else 0.3 for edge in G.edges()]
nx.draw(G, pos,
        with_labels=True,
        node_color=node_colors,
        node_size=2000,
        font_size=10,
        font_color='black',
        edge_color=edge_colors,
        width=edge_widths,
        alpha=0.8,
        arrows=True,
        arrowstyle='->',
        arrowsize=10
       )

plt.title('Bank Transaction Network (Highlighting Fraud Channels)', fontsize=20)
print("\nDisplaying network graph...")
print(" - Red Nodes: Banks directly involved in a fraudulent transaction.")
print(" - SkyBlue Nodes: Banks transacting with fraud-related banks.")
print(" - Red Edges: Confirmed fraudulent transactions between banks.")
print(" - Gray Edges: Legitimate transactions.")
plt.show()
```
